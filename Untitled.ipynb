```python
import numpy as np
import pickle
import mne
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# CSP
from mne.decoding import CSP

csp = CSP(n_components=4, norm_trace=False)

data1 = np.load(f'Topo_data/s01-LVLA.npy').astype('float64')
data2 = np.load(f'Topo_data/s01-HVLA.npy').astype('float64')
data3 = np.load(f'Topo_data/s01-LVHA.npy').astype('float64')
data4 = np.load(f'Topo_data/s01-HVHA.npy').astype('float64')
print(data2.shape)
X = np.stack((data1, data2, data3, data4), axis=0)
y = np.array([0,1,2,3])  # 0和1表示两个不同的类别

# 训练CSP
X_new = np.zeros((4, 32, 4))
for chan in range(32):
    tmp=X[:,chan,:][:, np.newaxis,:]
    # print(tmp)
    X_new[:,chan,:] = csp.fit_transform(tmp, y)
print(X_new.shape)
# 转换数据
# transformed_data_class1 = csp.transform(data1[np.newaxis, :, :])
# transformed_data_class2 = csp.transform(data2[np.newaxis, :,:])
# info = mne.create_info(ch_names=ch_names, sfreq=128., ch_types='eeg')
# evoked = mne.EvokedArray(transformed_data_class2.T, info)
# evoked.set_montage(biosemi_montage)
# csp.plot_patterns(evoked.info, ch_type="eeg", units="Patterns (AU)", size=1.5)
# print(transformed_data_class1)
```

```python
biosemi_montage = mne.channels.make_standard_montage('standard_1020')
ch_names = ['Fp1', 'AF3', 'F7', 'F3', 'FC1', 'FC5', 'T7', 'C3', 'CP1', 'CP5',
    'P7', 'P3', 'Pz', 'PO3', 'O1', 'Oz', 'O2', 'PO4', 'P4', 'P8',
    'CP6', 'CP2', 'C4', 'T8', 'FC6', 'FC2', 'F4', 'F8', 'AF4', 'Fp2',
    'Fz', 'Cz']
# band_list=[r'$\theta$', r'$\alpha$', r'$\beta$' , r'$\gamma$']
fig, axes = plt.subplots(figsize=(16, 10), nrows=3, ncols=4)
# fig.suptitle('CompareALL')# 'HVLA','LVHA',
classes = ['LVLA', 'HVLA','LVHA','HVHA']
# 找出所有数据中的最小值和最大值
# all_data = [np.load(f'Topo_data/s04-{classes[classe]}.npy') for classe in range(4)]
# all_data_means = [np.mean(data, axis=1) for data in all_data]
# vmin = min([data.min() for data in all_data_means])
# vmax = max([data.max() for data in all_data_means])
vmin = np.min(X_new)
vmax = np.max(X_new)

print('s01-CSPO')
for classe in range(4):
    # data = np.load(f'Topo_data/s04-{classes[classe]}.npy')
    # pca = PCA(n_components=3)
    # print(pca.fit(data).explained_variance_ratio_)
    
    # transformed_data = pca.fit_transform(data)
    # transformed_data = np.sum(transformed_data, axis=1)
    # data_sum = transformed_data.reshape(32,1)
    info = mne.create_info(ch_names=ch_names, sfreq=128., ch_types='eeg')
    evoked = mne.EvokedArray(X_new[classe,:, :], info)
    evoked.set_montage(biosemi_montage)
    for com in range(3):
        ax=axes[com, classe]
        im, _ = mne.viz.plot_topomap(evoked.data[:,com], evoked.info, show=False, axes=ax,
                         ch_type='eeg', cmap='Spectral_r', contours=10,vlim=(vmin, vmax),)#vlim=(vmin, vmax),np.mean(evoked.data, axis=1)
        ax.set_title(f'{classes[classe]}', fontsize=10)

# fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.3)
# fig.tight_layout()
plt.show()
```
